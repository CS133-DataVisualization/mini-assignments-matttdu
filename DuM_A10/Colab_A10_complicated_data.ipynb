### Question 1

blast_results: https://raw.githubusercontent.com/csbfx/advpy122-data/master/blast_results.csv

Read in the data from the csv file above. Skip all the comment lines, create a header for the dataframe based on the fields list in the comment line that starts with **# Fields:**. Drop the first column `query acc.ver`. Set the `subject acc.ver` as the index of the dataframe. The last column of the dataframe should be `publications`. Show the first five rows of the resulting dataframe.

```python
### Your code here . . .
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

header = ["query acc.ver", "subject acc.ver", "% identity", "alignment length", "mismatches", "gap opens", "q. start",
          "q. end", "s. start", "s. end", "evalue", "bit score", "% positives", "bonds", "publications"]

blast_results = pd.read_csv("https://raw.githubusercontent.com/csbfx/advpy122-data/master/blast_results.csv",
                            skiprows = 7, names = header, usecols = header, na_values = ['-'], thousands = ',')

blast_results = blast_results.drop(columns = ['query acc.ver']).set_index(['subject acc.ver'])
blast_results.head(5)
```

### Question 2
What is the average number of publications?

```python
### Your code here . . .
blast_results['publications'].mean()
```

### Question 3
List the `subject acc.ver` that has over 15,000 bonds.

```python
### Your code here . . .
bonds_range = pd.DataFrame(blast_results[blast_results['bonds'] > 15000].index)
bonds_range
```

### Question 4
Create a plot that shows the correlation between `% identity` and `% positives`.

```python
### Your code here . . .
sns.relplot(
    data = blast_results,
    x = '% identity',
    y = '% positives'
)

plt.title("% identity increases with % positives");
```

### Question  5
Add a column `Protein source` to the dataframe from Question 1 based on the data in this csv file: https://raw.githubusercontent.com/csbfx/advpy122-data/master/protein_source.csv. Merge the data when the `subject acc.ver` from the dataframe equals to `Protein` in the csv file. Entries without a matching protein will have `NA` as `Protein source`.

```python
### Your code here . . .
protein_data = pd.read_csv("https://raw.githubusercontent.com/csbfx/advpy122-data/master/protein_source.csv")
protein_data = protein_data.set_index('Protein')
protein_data = protein_data.drop(columns = 'Unnamed: 0')

blast_results['Protein source'] = protein_data
blast_results['Protein source'] = blast_results['Protein source'].fillna("NA")
blast_results
```
